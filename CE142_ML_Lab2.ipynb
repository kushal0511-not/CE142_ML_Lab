**FEATURE SELECTION & CORRELEATION**

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

Loading the dataset

```python
data = pd.read_csv('/Exercise-CarData.csv')
```

```python
data.head()
```

Removing the Class Label entry (Weight)

```python
data = data.iloc[:,:-1]
data.head()
```

```python
data.info()
```

Selecting features based on correlation

```python
corr = data.corr()
corr.head()
```

Generating the correlation heatmap

```python
sns.heatmap(corr)
```

compare the correlation between features and remove one of two features that have a correlation higher than 0.9

```python
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
```

```python
selected_columns = data.columns
selected_columns.shape
data = data[selected_columns]
print(data)
```
